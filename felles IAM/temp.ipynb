```python
import pandas as pd
```

```python
data = pd.read_csv("temp.csv", sep=";")
```

```python
data
```

```python
# This will only apply to columns with string (object) dtype
for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].str.replace('^Slått på', 'På', regex=True)
```

```python
data['Kontoadministrator'] = data['Kontoadministrator'].str.replace('^Slått på', "På", regex=True)
```

```python
data.replace(to_replace=r"^Slått på.*", value="På", regex=True, inplace=True)
data.replace(to_replace=r"^Deaktivert.*", value="Av", regex=True, inplace=True)
```

```python
data.to_csv("nye_data.csv")
```
