Først henter eg inn alle "web_logs-" som ligg i mappa:

```python
import pandas as pd
import os
# list all files matching "web_logs" in current directory
files = [f for f in os.listdir('.') if 'web_logs-2025' in f]
# read all files into a single dataframe 
data = pd.concat([pd.read_csv(f, dtype=str) for f in files])
```

```python
data.info()
```

Så tar eg vekk ein del rader frå matrisa:

- alle som har "/api/" i url
- alle utan *user_id*
- alle utan url i det heile (men eg må ta vare på dei med "/", for det er dashbordet til brukaren)

```python
t1 = data[~data['value.url'].str.contains("/api/")]
utan_user_id = t1[t1['value.user_id'].isna()]
med_user_id = t1[~t1['value.user_id'].isna()]
```

```python
data_til_bruk = med_user_id[['value.timestamp', 'value.user_id', 'value.course_id', 'value.web_application_controller', 'value.web_application_action', 'value.web_application_context_type', 'value.url']]
data_til_bruk.to_csv('data_til_bruk.csv', index=False)
```
