```python
import pandas as pd
```

```python
sider = pd.read_csv("sider.csv")
sider.info()
```

```python
def splitt(rad):
    if '?' in rad:
        return rad.split('?')[0]
    else:
        return rad
```

```python
sider['strippa'] = sider['index'].apply(splitt)
```

```python
sider
```

```python
mest_sett = sider.groupby('emne').sum().sort_values('frekvens', ascending=False)
```

```python
for i, r in sider[sider['emne']==mest_sett.index[0]].iterrows():
    print(r['strippa'], r['frekvens'])
```

```python
liste = sider[sider['emne']==mest_sett.index[0]].groupby('strippa').sum().sort_values('frekvens', ascending=False)
```

```python
print("Besøkte sider:")
for i, r in liste.iterrows():
    if '/edit' not in i:
        print(f"{r['frekvens']:3} besøk: {i}")
print("Redigerte sider:")
for i, r in liste.iterrows():
    if '/edit' in i:
        print(f"{r['frekvens']:3} redigeringer: {i[0:-5]}")
```
